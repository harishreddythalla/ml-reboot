```python
import numpy as np
import matplotlib.pyplot as plt

#Generating synthetic data
np.random.seed(42)
X = np.linspace(0,10,50) #creating 50 evenly distributed data points between 0 and 10
y = 3*X+7+np.random.randn(50)*2

plt.scatter(X,y)
plt.title("Synthetic Data")
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```

```python
#Initializing the parameters of our Linear Regression equation y= mx+b

m = 0.0 #slope
c = 0.0 #intercept
learning_rate = 0.01
epochs = 1000


#loss function
def mse_loss(y, y_pred):
    return np.mean((y-y_pred)**2)

n = len(X)
#Gradient Descent
for epoch in range(epochs):
    y_pred = m*X+c
    mse = mse_loss(y, y_pred)
    dm = (-2/n)* np.sum(X*(y-y_pred))
    dc = (-2/n)* np.sum(y-y_pred)
    m -= learning_rate*dm
    c -= learning_rate*dc
    if epoch%100 ==0:
        print(f"Epoch {epoch}: m={m:.3f}, c={c:.3f}, loss={mse_loss(y, y_pred):.3f}")
```

```python
y_final = m*X+c
plt.scatter(X,y)
plt.plot(X, y_final, color = 'red')
plt.title('Fitted Line')
plt.show()
```

```python
import json

weights = {"m": m, "c": c}
with open("linear_model.json", "w") as f:
    json.dump(weights, f)
```
